# 2. Estimate pore size distribution

## 2-1. When the spacing between the pore diameters was 0.2 angstrom

### 2-1-1. Compute the relative errors between each estimated value and experimental data 

```python
pore_range_2step = []
N2_SL_range_2step = []
N2_BET_range_2step = []
AV_range_2step = []

for rrre in np.arange(3.9,7.5,0.2):
    
    N2_SL_value_2step = linear_N2_SL_three(rrre)
    N2_BET_value_2step = linear_BET_three(rrre)
    AV_value_2step = linear_AV_three(rrre)
    
    pore_range_2step.append(rrre)
    
    N2_SL_range_2step.append(N2_SL_value_2step)
    N2_BET_range_2step.append(N2_BET_value_2step)
    AV_range_2step.append(AV_value_2step)

N2_SL_error_DAM_2step_list = []
N2_SL_error_DABA_2step_list = []
N2_SL_error_BPDA_2step_list = []

N2_BET_error_DAM_2step_list = []
N2_BET_error_DABA_2step_list = []
N2_BET_error_BPDA_2step_list = []

AV_error_DAM_2step_list = []
AV_error_DABA_2step_list = []
AV_error_BPDA_2step_list = []

for mmme in N2_SL_range_2step:
    
    N2_SL_error_DAM_2step = np.abs((mmme-N2_SL_exp[0])/N2_SL_exp[0])*100
    N2_SL_error_DAM_2step_list.append(N2_SL_error_DAM_2step)
    
    N2_SL_error_DABA_2step = np.abs((mmme-N2_SL_exp[1])/N2_SL_exp[1])*100
    N2_SL_error_DABA_2step_list.append(N2_SL_error_DABA_2step)
    
    N2_SL_error_BPDA_2step = np.abs((mmme-N2_SL_exp[2])/N2_SL_exp[2])*100
    N2_SL_error_BPDA_2step_list.append(N2_SL_error_BPDA_2step)
    
for mmee in N2_BET_range_2step:
    
    N2_BET_error_DAM_2step = np.abs((mmee-N2_BET_exp[0])/N2_BET_exp[0])*100
    N2_BET_error_DAM_2step_list.append(N2_BET_error_DAM_2step)
    
    N2_BET_error_DABA_2step = np.abs((mmee-N2_BET_exp[1])/N2_BET_exp[1])*100
    N2_BET_error_DABA_2step_list.append(N2_BET_error_DABA_2step)
    
    N2_BET_error_BPDA_2step = np.abs((mmee-N2_BET_exp[2])/N2_BET_exp[2])*100
    N2_BET_error_BPDA_2step_list.append(N2_BET_error_BPDA_2step)

for tax in AV_range_2step:
    
    AV_error_DAM_2step = np.abs((tax-AV_exp[0])/AV_exp[0])*100
    AV_error_DAM_2step_list.append(AV_error_DAM_2step)
    
    AV_error_DABA_2step = np.abs((tax-AV_exp[1])/AV_exp[1])*100
    AV_error_DABA_2step_list.append(AV_error_DABA_2step)
    
    AV_error_BPDA_2step = np.abs((tax-AV_exp[2])/AV_exp[2])*100
    AV_error_BPDA_2step_list.append(AV_error_BPDA_2step)
```

```python
pore_range_5step = []
N2_SL_range_5step = []
N2_BET_range_5step = []
AV_range_5step = []

for rrrrree in np.arange(3.9,7.9,0.5):
    
    N2_SL_value_5step = linear_N2_SL_three(rrrrree)
    N2_BET_value_5step = linear_BET_three(rrrrree)
    AV_value_5step = linear_AV_three(rrrrree)
    
    pore_range_5step.append(rrrrree)
    
    N2_SL_range_5step.append(N2_SL_value_5step)
    N2_BET_range_5step.append(N2_BET_value_5step)
    AV_range_5step.append(AV_value_5step)

N2_SL_error_DAM_5step_list = []
N2_SL_error_DABA_5step_list = []
N2_SL_error_BPDA_5step_list = []

N2_BET_error_DAM_5step_list = []
N2_BET_error_DABA_5step_list = []
N2_BET_error_BPDA_5step_list = []

AV_error_DAM_5step_list = []
AV_error_DABA_5step_list = []
AV_error_BPDA_5step_list = []

for mmmeeee in N2_SL_range_5step:
    
    N2_SL_error_DAM_5step = np.abs((mmmeeee-N2_SL_exp[0])/N2_SL_exp[0])*100
    N2_SL_error_DAM_5step_list.append(N2_SL_error_DAM_5step)
    
    N2_SL_error_DABA_5step = np.abs((mmmeeee-N2_SL_exp[1])/N2_SL_exp[1])*100
    N2_SL_error_DABA_5step_list.append(N2_SL_error_DABA_5step)
    
    N2_SL_error_BPDA_5step = np.abs((mmmeeee-N2_SL_exp[2])/N2_SL_exp[2])*100
    N2_SL_error_BPDA_5step_list.append(N2_SL_error_BPDA_5step)
    
for mmeeeee in N2_BET_range_5step:
    
    N2_BET_error_DAM_5step = np.abs((mmeeeee-N2_BET_exp[0])/N2_BET_exp[0])*100
    N2_BET_error_DAM_5step_list.append(N2_BET_error_DAM_5step)
    
    N2_BET_error_DABA_5step = np.abs((mmeeeee-N2_BET_exp[1])/N2_BET_exp[1])*100
    N2_BET_error_DABA_5step_list.append(N2_BET_error_DABA_5step)
    
    N2_BET_error_BPDA_5step = np.abs((mmeeeee-N2_BET_exp[2])/N2_BET_exp[2])*100
    N2_BET_error_BPDA_5step_list.append(N2_BET_error_BPDA_5step)

for taxeee in AV_range_5step:
    
    AV_error_DAM_5step = np.abs((taxeee-AV_exp[0])/AV_exp[0])*100
    AV_error_DAM_5step_list.append(AV_error_DAM_5step)
    
    AV_error_DABA_5step = np.abs((taxeee-AV_exp[1])/AV_exp[1])*100
    AV_error_DABA_5step_list.append(AV_error_DABA_5step)
    
    AV_error_BPDA_5step = np.abs((taxeee-AV_exp[2])/AV_exp[2])*100
    AV_error_BPDA_5step_list.append(AV_error_BPDA_5step)
```

### 2-4-2. Estimate pore size distribution for DAM sample

```python
# create a LP minimization problem
Lp_prob_DAM_5step = p.LpProblem('Problem', p.LpMinimize)

# create problem variables
x1 = p.LpVariable("x1", lowBound = 0)
x2 = p.LpVariable("x2", lowBound = 0)
x3 = p.LpVariable("x3", lowBound = 0)
x4 = p.LpVariable("x4", lowBound = 0)
x5 = p.LpVariable("x5", lowBound = 0)
x6 = p.LpVariable("x6", lowBound = 0)
x7 = p.LpVariable("x7", lowBound = 0)

variable_list_DAM_5step=[x1,x2,x3,x4,x5,x6,x7,
                         1-x1-x2-x3-x4-x5-x6-x7]

Sum_AV_SL_BET_error_DAM_5step=0
Sum_SL_DAM_5step=0
Sum_BET_DAM_5step=0
Sum_AV_DAM_5step=0

for nnnnnne in range(len(variable_list_DAM_5step)):
    
    aaaaaae=(N2_SL_error_DAM_5step_list[nnnnnne]+N2_BET_error_DAM_5step_list[nnnnnne]+AV_error_DAM_5step_list[nnnnnne])*variable_list_DAM_5step[nnnnnne]
    Sum_AV_SL_BET_error_DAM_5step=Sum_AV_SL_BET_error_DAM_5step+aaaaaae

    bbbbbbe = N2_SL_range_5step[nnnnnne]*variable_list_DAM_5step[nnnnnne]
    Sum_SL_DAM_5step=Sum_SL_DAM_5step+bbbbbbe
    
    cccccce = N2_BET_range_5step[nnnnnne]*variable_list_DAM_5step[nnnnnne]
    Sum_BET_DAM_5step = Sum_BET_DAM_5step+cccccce
    
    cccchoco = AV_range_5step[nnnnnne]*variable_list_DAM_5step[nnnnnne]
    Sum_AV_DAM_5step = Sum_AV_DAM_5step + cccchoco

print(Sum_AV_SL_BET_error_DAM_5step) 

# objective function 
Lp_prob_DAM_5step += Sum_AV_SL_BET_error_DAM_5step

# mixing constraints
Lp_prob_DAM_5step += (1-x1-x2-x3-x4-x5-x6-x7) >= 0
Lp_prob_DAM_5step += Sum_BET_DAM_5step >= N2_BET_exp[0]*0.94
Lp_prob_DAM_5step += Sum_BET_DAM_5step <= N2_BET_exp[0]*1.06
Lp_prob_DAM_5step += Sum_AV_DAM_5step >= AV_exp[0]*0.94
Lp_prob_DAM_5step += Sum_AV_DAM_5step <= AV_exp[0]*1.06
Lp_prob_DAM_5step += Sum_SL_DAM_5step >= N2_SL_exp[0]*0.94
Lp_prob_DAM_5step += Sum_SL_DAM_5step <= N2_SL_exp[0]*1.06

# display the problem
print(Lp_prob_DAM_5step)

status_DAM_5step = Lp_prob_DAM_5step.solve()
print(p.LpStatus[status_DAM_5step])

final_variable_list_DAM_5step =[p.value(x1),p.value(x2),p.value(x3),p.value(x4),p.value(x5),p.value(x6),p.value(x7),
                                p.value(1-x1-x2-x3-x4-x5-x6-x7)]

fig=plt. figure(figsize =(13,10)) 
ax1613=fig.add_subplot(1,1,1)
plt. xlim(3.5,7.5)
plt. ylim(0,1.1)
plt. xticks(fontsize=28)
plt. tick_params(axis='both', width=4, length=10)
plt. yticks(fontsize=28)
for axis in ['top','bottom','left','right']:
    ax1613.spines[axis].set_linewidth(4)

plt.bar(pore_range_5step, final_variable_list_DAM_5step, color='blue', alpha=0.5, width=0.1, label='DAM')

plt.xlabel('Pore Diameter (Å)', fontsize=36)
plt.ylabel('Normalized Frequency (-)', fontsize=36)
plt.legend(frameon=True, loc = 'best',fontsize=24)
plt. savefig("./DAM_three_all_6_0928_05step.png",format='png')
plt.show()

for ooooooe in range(len(final_variable_list_DAM_5step)):
    print(pore_range_5step[ooooooe],final_variable_list_DAM_5step[ooooooe])

sum_N2_SL_final_DAM_5step=0
sum_N2_BET_final_DAM_5step=0
sum_AV_final_DAM_5step=0

for kkkkkke in range(len(final_variable_list_DAM_5step)):
    dddcdde = N2_SL_range_5step[kkkkkke]*final_variable_list_DAM_5step[kkkkkke]
    sum_N2_SL_final_DAM_5step=sum_N2_SL_final_DAM_5step+dddcdde

    ddddragon = N2_BET_range_5step[kkkkkke]*final_variable_list_DAM_5step[kkkkkke]
    sum_N2_BET_final_DAM_5step=sum_N2_BET_final_DAM_5step+ddddragon
    ssssupport = AV_range_5step[kkkkkke]*final_variable_list_DAM_5step[kkkkkke]
    sum_AV_final_DAM_5step=sum_AV_final_DAM_5step+ssssupport

ddddddee=np.abs((sum_N2_SL_final_DAM_5step-N2_SL_exp[0])/N2_SL_exp[0])*100  
aaaapple=np.abs((sum_N2_BET_final_DAM_5step-N2_BET_exp[0])/N2_BET_exp[0])*100
bbbbanana=np.abs((sum_AV_final_DAM_5step-AV_exp[0])/AV_exp[0])*100

print("N2 SL for DAM = "+str(sum_N2_SL_final_DAM_5step))
print("error (%) ="+str(ddddddee))
print("N2 BET for DAM = "+str(sum_N2_BET_final_DAM_5step))
print("error (%)"+str(aaaapple))
print("AV for DAM = "+str(sum_AV_final_DAM_5step))
print("error (%)"+str(bbbbanana))
```
